# product date i doldurmak

```python
for i, row in df903.iterrows():
    # first_registration sütunundaki tarihi parse et
    date = pd.to_datetime(row['first_registration'], format='%m/%Y')
    year, month = date.year, date.month
    
    # Yılın ilk yarısı için
    if month <= 6:
        # production_date sütununa bir önceki yılın bilgisini yaz
        df903.at[i, 'production_date'] = pd.to_datetime(f"{year-1}", format='%Y')
    # Yılın ikinci yarısı için
    else:
        # production_date sütununa o yılın bilgisini yaz
        df903.at[i, 'production_date'] = pd.to_datetime(f"{year}", format='%Y')
# "production_date" sütununu datetime tipine dönüştürelim
df903['production_date'] = pd.to_datetime(df903['production_date'])
```
