```python
import pandas as pd
import json

# JSON dosyasını yükleyin
with open('as24_cars.json', 'r') as f:
    data = json.load(f)

# DataFrame'e dönüştürün
df0 = pd.json_normalize(data)

# DataFrame'i yazdırın
print(df0)
```

```python
df = df0.copy()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# make_model sütununu make ve model sütunlarına ayırma
df[['make', 'model']] = df['make_model'].str.split(' ', 1, expand=True)
```

```python
# make_model sütununu kaldırma
df.drop('make_model', axis=1, inplace=True)
```

```python
df.head(1)
```

```python
df.info()
```

```python
df = df.replace('\n','', regex=True).applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df.head(10)
```

```python
for group1 in df ["make_model"].unique():
    for group2 in list (df["body_type"].unique()):
        cond2 = (df["make_model"]==group1) & (df["body_type"]==group2)
        df.loc[cond2, "Paint_Type"] = df.loc[cond2, "Paint_Type"].fillna (method="ffill").fillna(method="bfill")
for group1 in list (df["make_model"].unique()):
    cond1 = df["make_model"]==group1
    df.loc[cond1, "Paint_Type"] = df.loc[cond1, "Paint_Type"].fillna (method="ffill").fillna (method="bfill")
df["Paint_Type"] = df["Paint_Type"].fillna (method="ffill").fillna (method="bfill")
```
